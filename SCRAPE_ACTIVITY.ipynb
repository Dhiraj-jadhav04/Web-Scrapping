**Reasoning**:
After reading the HTML content, the next logical step is to parse it using BeautifulSoup, extract all quotes, and then process each quote to find those tagged with 'life', similar to how it was done for a single file previously.



```python
import requests
from bs4 import BeautifulSoup

for page_number in range(1, 11):
  file_path = f"scraped_data{page_number}.html"
  with open(file_path, "r") as f:
    html_content = f.read()

  soup = BeautifulSoup(html_content, "lxml")
  all_quotes = soup.select("div.quote")

  for q in all_quotes:
    all_tags = []
    for tag in q.select(".tags .tag"):
      all_tags.append(tag.get_text())

    if "life" in all_tags:
      text = q.select_one("span.text").get_text()
      author = q.select_one("small.author").get_text()
      all_life_quotes.append({"text": text, "author": author})

print("Extracted all 'life' quotes from all pages and aggregated them.")
```
